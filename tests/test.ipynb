```python
import math
```

```python
start_lambda = 0.8
end_lambda = 0.4
start_iter = 200
end_iter = 800

def cosine_lambda(iter_, start_lambda, end_lambda, start_iter, end_iter):
    if iter_ < start_iter:
        return start_lambda
    if iter_ > end_iter:
        return end_lambda
    
    num_total_iters = end_iter - start_iter
    iter_ -= start_iter
    lambda_ = 0.5 * (math.cos(iter_ / num_total_iters * math.pi) + 1)
    lambda_ = lambda_ * (start_lambda - end_lambda) + end_lambda
    return lambda_

import matplotlib.pyplot as plt
plt.plot([cosine_lambda(iter_, start_lambda, end_lambda, start_iter, end_iter) for iter_ in range(0, 1000 + 1, 10)])
```
